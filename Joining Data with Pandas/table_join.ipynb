```python
import pandas as pd
wards = pd.read_pickle('Data_files/ward.p')
wards.head() 
```

```python
census = pd.read_pickle('Data_files/census.p')
census.head()
```

```python
#merging census and wards table
ward_census = wards.merge(census, on='ward')
ward_census.head()
```

```python
ward_census2 = wards.merge(census, on='ward', suffixes=('_ward','_cen'))
ward_census2.head()
ward_census2.shape
```

```python
licenses = pd.read_pickle('Data_files/licenses.p')
licenses.head()
```

```python
#Merging multiple tables together
mul_merge = licenses.merge(wards, on=['address', 'zip'])
mul_merge.head()
```
